### 5.2 Libraries that will be used<a id='5.2_Libraries_that_will_be_used'></a>

```python
import pandas as pd
import lightgbm as lgb
import pickle

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from pandas import MultiIndex, Int16Dtype

```

### 5.4 Loading the data<a id='5.4_Loading_the_data'></a>

```python
# Load saved data Step_3 EDA.
df = pd.read_csv('/Users/aquinojoeanson/Desktop/SPRINGBOARD/Capstone_Project_2/Data/Cleveland_Model.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

### 5.5 Logistic Regression Classifier<a id='5.5_logisticregression'></a>

```python
Xlr, Xtestlr, ylr, ytestlr = train_test_split(X, y, random_state = 2, stratify=y)

clf = LogisticRegression(solver = 'newton-cg', max_iter = 500)

clf.fit(Xlr, ylr)

# Predicting the Test set results
y_predict_test = clf.predict(Xtestlr)

#Pickle
newfilename_pickle = 'cp_predict_model'
pickle.dump(clf,open(newfilename_pickle,'wb'))

cm_test = confusion_matrix(y_predict_test, ytestlr)

loaded_model = pickle.load(open(newfilename_pickle,'rb'))
loaded_model.predict(Xtestlr)

# Predicting the Train set results
y_predict_train = clf.predict(Xlr)
cm_train = confusion_matrix(y_predict_train, ylr)

#print(newfilename_pickle)
#print(cm_test)
#print(cm_train)

print(classification_report(y_predict_test, ytestlr))
print(classification_report(y_predict_train, ylr))
print('Accuracy for training set for Logistic Regression = {}'.format((cm_train[0][0] + cm_train[1][1])/len(ylr)))
print('Accuracy for test set for Logistic Regression = {}'.format((cm_test[0][0] + cm_test[1][1])/len(ytestlr)))
```

### 5.6 Random Forest Classifier <a id='5.6_randomforest'></a>

```python
Xlr, Xtestlr, ylr, ytestlr = train_test_split(X, y, random_state = 2, stratify=y)

clf = RandomForestClassifier(n_estimators = 10)
clf.fit(Xlr, ylr)

# Predicting the Test set results
y_predict_test = clf.predict(Xtestlr)

#Pickle
newfilename_pickle = 'cp_predict_model'
pickle.dump(clf,open(newfilename_pickle,'wb'))

cm_test = confusion_matrix(y_predict_test, ytestlr)

loaded_model = pickle.load(open(newfilename_pickle,'rb'))
loaded_model.predict(Xtestlr)

#from sklearn.metrics import confusion_matrix
#cm_test = confusion_matrix(y_predict_test, ytestlr)

y_predict_train = clf.predict(Xlr)
cm_train = confusion_matrix(y_predict_train, ylr)

print(classification_report(y_predict_test, ytestlr))
print(classification_report(y_predict_train, ylr))
print('Accuracy for training set for Random Forest = {}'.format((cm_train[0][0] + cm_train[1][1])/len(ylr)))
print('Accuracy for test set for Random Forest = {}'.format((cm_test[0][0] + cm_test[1][1])/len(ytestlr)))
```

### 5.7 LightGBM Classifier<a id='5.7_lgbm'></a>

```python
d_train = lgb.Dataset(Xlr, label = ylr)
params = {}

clf = lgb.train(params, d_train, 100)

#Prediction
y_predict_test = clf.predict(Xtestlr)



#convert into binary values
for i in range(0, len(y_predict_test)):
    if y_predict_test[i]>= 0.5:       # setting threshold to .5
       y_predict_test[i]=1
    else:  
       y_predict_test[i]=0
       
cm_test = confusion_matrix(y_predict_test, ytestlr)

#Pickle
newfilename_pickle = 'cp_predict_model'
pickle.dump(clf,open(newfilename_pickle,'wb'))

cm_test = confusion_matrix(y_predict_test, ytestlr)

loaded_model = pickle.load(open(newfilename_pickle,'rb'))
loaded_model.predict(Xtestlr)
y_predict_train = clf.predict(Xlr)

for i in range(0, len(y_predict_train)):
    if y_predict_train[i]>= 0.5:       # setting threshold to .5
       y_predict_train[i]=1
    else:  
       y_predict_train[i]=0
       
cm_train = confusion_matrix(y_predict_train, ylr)

print(classification_report(y_predict_test, ytestlr))
print(classification_report(y_predict_train, ylr))
print('Accuracy for training set for LightGBM = {}'.format((cm_train[0][0] + cm_train[1][1])/len(ylr)))
print('Accuracy for test set for LightGBM = {}'.format((cm_test[0][0] + cm_test[1][1])/len(ytestlr)))
```

### 5.8 xgBoost Classifier<a id='5.8_xgboost'></a>

```python
xg = XGBClassifier(eval_metric='mlogloss', use_label_encoder=False)
xg.fit(Xlr, ylr)
y_predict_test = xg.predict(Xtestlr)

cm_test = confusion_matrix(y_predict_test, ytestlr)

y_predict_train = xg.predict(Xlr)

for i in range(0, len(y_predict_train)):
    if y_predict_train[i]>= 0.5:       # setting threshold to .5
       y_predict_train[i]=1
    else:  
       y_predict_train[i]=0
    
#Pickle
newfilename_pickle = 'cp_predict_model'
pickle.dump(clf,open(newfilename_pickle,'wb'))

cm_test = confusion_matrix(y_predict_test, ytestlr)

loaded_model = pickle.load(open(newfilename_pickle,'rb'))
loaded_model.predict(Xtestlr)
cm_train = confusion_matrix(y_predict_train, ylr)

print(classification_report(y_predict_test, ytestlr))
print(classification_report(y_predict_train, ylr))
print('Accuracy for training set for XGBoost = {}'.format((cm_train[0][0] + cm_train[1][1])/len(ylr)))
print('Accuracy for test set for XGBoost = {}'.format((cm_test[0][0] + cm_test[1][1])/len(ytestlr)))
```
